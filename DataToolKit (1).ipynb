## Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import bokeh.io
import bokeh.plotting
import numpy as np
from bokeh.plotting import figure,show
bokeh.io.output_notebook()
x=np.arange(0,10)
y=np.sin(x)
p=figure(title="Wave chart ")
p.xaxis.axis_label="Set x-values from 0 to 10"
p.yaxis.axis_label="y-values as the sine of x"
p.line(x=x,y=y,color="orange")
show(p)
```

## Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset  
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']   
counts = [20, 25, 30, 35]

```python
from bokeh.io import output_notebook
import bokeh.plotting
from bokeh.plotting import figure,show,output_file
from bokeh.models import ColumnDataSource
bokeh.io.output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Prepare the data source 
source=ColumnDataSource(data=dict(fruits=fruits,counts=counts))
fig=figure(x_range=fruits,x_axis_label='Fruit',y_axis_label="Count",width=800,height=400,title="Generate a Bokeh bar chart representing the counts of different fruits using the following dataset")
fig.vbar(x='fruits',top='counts',width=0.5,source=source,color="red",alpha=0.7)

show(fig)
```

## Q4. Create a Bokeh histogram to visualize the distribution of the given data.  
data_hist = np.random.randn(1000)   
hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.io import output_notebook
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure,show,output_file

output_file("test.html")

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p=figure(title=" Bokeh histogram to visualize the distribution of the given data",width=600, height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="olive", line_color="white",
         legend_label="1000 random samples")

show(p)
```

## Q5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)  
x = np.linspace(0, 1, 10)  
y = np.linspace(0, 1, 10)   
xx, yy = np.meshgrid(x, y)


```python
from bokeh.io import output_notebook
import bokeh.plotting
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar
bokeh.io.output_notebook()
import numpy as np

output_file("test.html")

#data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat=xx.flatten()
y_flat=yy.flatten()
values = data_heatmap.flatten()
mapper = linear_cmap(field_name='value',palette=Viridis256 ,low=min(values) , high=max(values))


fig=figure(title="Heatmap",x_axis_label='X' , y_axis_label="Y",width=600,height=600)
fig.rect(x='x',y='y',width=0.1, height=0.1, source=dict(x=x_flat,y=y_flat,value=values),
         fill_color=mapper , line_color=None)
color_bar=ColorBar(color_mapper=mapper['transform'],width=8,location=(0,0))
fig.add_layout(color_bar,'right')

show(fig)




```
